# <font color="Yellow"> **Conditional Statement** </font>

##<font color="lime">**One-way if Statements:**</font> A one-way if statement executes the statements if the condition is true.
Capture.PNG



```python
iA = int(input("Enter an integer: "))
if iA % 5 == 0:
    print("🍏")
else:
    print("🍅")

print("Hello Friends!!!")
```

## <font color="lime"> Two-Way if-else Statements: </font>
A two-way if-else statement decides which statements to execute based on whether
the condition is true or false.

Capture.PNG

```python
iA = int(input("Enter an integer: "))
if iA % 2 == 0:
    print(iA, "is an even number.")
else:
    print(iA, "is an odd number.")
```

## <font color="lime"> Nested if and Multi-Way if-elif-else Statements:</font>

One <font color="aqua">*if*</font> statement can be placed inside another <font color="aqua">*if*</font> statement to form a nested <font color="aqua">*if* </font>statement.

Capture.PNG


```python
fA = float(input("Enter score: "))
if fA > 100.0:
    grade = "Invalid"
elif fA >= 90.0:
    grade = 'A'
elif fA >= 80.0:
    grade = 'B'
elif fA >= 70.0:
    grade = 'C'
elif fA >= 60.0:
    grade = 'D'
else:
    grade = 'F'

print("Your Grade is:",grade)
```

# <font color="Yellow"> Logical (or) Boolean  Operators:
The logical operators <font color="aqua">not</font>, <font color="aqua">and</font>, and <font color="aqua">or</font> can be used to create a **composite condition**.

Capture.PNG

```python
## >>> Operator "not" <<< ##
age = 24
#print(age > 18)
#print(not (age > 18))

#print(age < 18)
print(not(age < 18))
```

```python
 ## >>> Operator "and" <<< ##
age = 24
gender = 'F'
print((age > 18) and (gender == 'F'))
print((age > 25) and (gender == 'F'))
print((age > 18) and (gender != 'F'))
print((age > 25) & (gender != 'F'))
```

```python
## >>> Operator "or" <<< ##
print((age > 18) or (gender == 'F'))
print((age > 25) or (gender == 'F'))
print((age > 18) | (gender != 'F'))
print((age > 25) | (gender != 'F'))
```

#<Font color = "Yellow"> Operator Precedence Chart

Capture.PNG



If operators with the same precedence are next to each other, their associativity determines
the order of evaluation. All binary operators are **left-associative**.


For example, since + and –
are of the same precedence and are left-associative, the expression

a - b + c – d

is equivalent to

((a - b) + c) - d
